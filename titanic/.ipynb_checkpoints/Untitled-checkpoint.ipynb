```python
#get median age for each Pclass and Gender
median_ages = np.zeros((2,3))
median_ages
```

```python
for i in range(0,2):
    for j in range(0,3):
        median_ages[i][j] = titanic.loc[((titanic['Gender']==i)&(titanic['Pclass']==j+1)),'Age'].dropna().median()
median_ages
```

```python
titanic['AgeFill'] = titanic['Age']
titanic[titanic['Age'].isnull()].head()
```

```python
for i in range(0,2):
    for j in range(0,3):
        titanic.loc[((titanic.Age.isnull())&(titanic.Gender==i)&(titanic.Pclass==j+1)),'AgeFill'] = median_ages[i][j]
titanic[titanic['Age'].isnull()].head()
```

```python
titanic['AgeIsNull'] = titanic['Age'].isnull().astype(int)
titanic[titanic.Age.isnull()==True].head()
```

```python
titanic.describe()
```

```python
titanic['FamilySize'] = titanic['SibSp'] + titanic['Parch']
```

```python
titanic['Age*Pclass'] = titanic['AgeFill'] * titanic['Pclass']
```

```python
titanic['Age*Pclass'].hist(bins=20)
plt.show()
```

```python
drop_list = titanic.dtypes[titanic.dtypes.map(lambda x:x=='object')]
```

```python
titanic_numbers = titanic.drop(drop_list.index,axis=1)
```

```python
titanic_numbers = titanic_numbers.drop('Age',axis=1)
titanic_numbers = titanic_numbers.drop('PassengerId',axis=1)
```

```python
titanic_train
```

```python
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier(n_estimators=100)
```

```python
forest = forest.fit(titanic_train[0::,1::],titanic_train[0::,0])
```
